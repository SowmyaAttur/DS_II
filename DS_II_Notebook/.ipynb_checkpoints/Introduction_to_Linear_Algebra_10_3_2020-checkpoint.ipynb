```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
np.arange(0,10,1)
```

```python
x=np.arange(0,10,.1)
```

```python
#function for a straight line
def f(x):
    return x
```

```python
y = f(x)
```

```python
plt.figure()
plt.plot(x,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
px.line(x,y,labels = {"x":"X-axis","index":"Y-axis"})
```

```python
#parabola
x2 = np.arange(-10000,10000,1)
    
```

```python
def f(x):
    return np.power(x,2)
```

```python
y = f(x2)
```

```python
px.line(x= x2,y =y,labels={"x":'x-Axis',"y":"y-axis"})
```

```python
def f(x):
    return np.power(x,2)
x2= np.arange(-10000,10000,1)
y = f(x2)
plt.figure
plt.xlabel("x2")
plt.ylabel("y")
plt.plot(x2, y)
```

```python
#exponential function
def f(x):
    return np.exp(x)
x2= np.arange(0,10,0.01)
y = f(x2)
plt.figure
plt.xlabel("x2")
plt.ylabel("y")
plt.plot(x2, y)
```

```python
#exponential function
def f(x):
    return 1/(1+np.exp(x))
x2= np.arange(0,10,0.01)
y = f(x2)
plt.figure
plt.xlabel("x2")
plt.ylabel("y")
plt.plot(x2, y)
```

```python
#logistic regression function
def f(x):
    return 1/(1+np.exp(-x))
x2= np.arange(-10,10,0.01)
y = f(x2)
plt.figure
plt.xlabel("x2")
plt.ylabel("y")
plt.plot(x2, y)
```
